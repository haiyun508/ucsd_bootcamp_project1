```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# read in csv file for data
all_data = pd.read_csv("Resources/all_data.csv")

# read in csv for lat/lng
coordinates = pd.read_csv("Resources/coords.csv", encoding="Latin-1")
```

```python
# groupby object for country and region
country_group_df = all_data.groupby(['Country','Region']).mean()
country_group_df = country_group_df.reset_index()
country_group_df.head()
# merge coordinates to dataframe
merged_df = country_group_df.merge(coordinates, how="left", left_on="Country", right_on="name")

# drop year column
merged_df = merged_df.drop(columns=["Year","name","country"])
merged_df.head()
```
